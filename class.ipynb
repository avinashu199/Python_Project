```python
'''
class is a blue print which is similar to module
moudle will have constants and functions
class will have variables and methods
To access things of modules we shall import them using directly
To access things of class we should create the object and use the object to access varibales and methods of that class
'''
```

```python
'''
in class there are 2 main things that need to be know
1.self-- refrence to the object that we are going to create
2.init constructor- This will assign the values to the object as soon as it is created
'''
```

```python
class pen:
    #All the varibales/instance variable should be created inside the constructor
    def __init__(self,name,color,price):
        self.name=name
        self.color=color
        self.price=price
    def description(self):
        print(f"This is a {self.name} pen of {self.color} color that costs ${self.price}")

obj=pen('Pinpoint','Blue',10)
obj.description()        
#object is having all the instance variables we can use them to access those values
print(obj.name,obj.color,obj.price)
```

```python
#Another example of class
class car:
    def __init__(self,brand,color,fuel,price):
        self.brand=brand
        self.color=color
        self.fuel=fuel
        self.price=price
    def desc(self):
        print(f"I have a {self.color} color {self.brand} car, running on {self.fuel} with an road price ${self.price}")
#create an object for car class
obj=car('Ford','Silver','Diesel','11Lakh')
obj.desc()
```
