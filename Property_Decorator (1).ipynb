```python
'''
Property decorator is used to acheive encapsulation in python
using this propery we can access the methods like attributes, so what we will do here using property attribute is we will create private instance variables
and we will use property method to access,modify and delete those varibales

normally in this scenario if we try to access the instance varibale, python will execute the method for that instance varible
'''
```

```python
# Calling methods like variables using property decorators

class Employee:
    def __init__(self,name,age,job_title,salary):
        self._name=name  #private instance varibales
        self._age=age #private instance varibales
        self._job_title=job_title #private instance varibales
        self.salary=salary #private instance varibales
    #now create the methods for those varibales so that python will call the methods instead of directly calling the variables
    @property
    def name(self):
        return self._name.upper()
    @property
    def age(self):
        if self._age<22:
            return f"{self._name} is a small boy of {self._age} years"
        else:
            return f"{self._name} is an adult of {self._age} years"
employee1=Employee('Avinash',31,'Sr.QA',85000)
employee2=Employee("Ayaansh",2,"Jr.QA",20000)
employee3=Employee("Kavita",24,"Police",40000)
print(employee1.name)
print(employee2.name)
print(employee3.name)
print(employee1.age)
print(employee2.age)
print(employee3.age)
```

```python
# setter and deleter functionality of the property decarotor
class Employee:
    def __init__(self,name,age,job_title,salary):
        self._name=name  #private instance varibales
        self._age=age #private instance varibales
        self._job_title=job_title #private instance varibales
        self._salary=salary #private instance varibales
    #now create the methods for those varibales so that python will call the methods instead of directly calling the variables
    @property #to read/access
    def name(self):
        return self._name.upper()
    @property #to read/access
    def age(self):
        return self._age
    @property #to read/access
    def job_title(self):
        return self._job_title
    @property #to read/access
    def salary(self):
        return self._salary
    @salary.setter #to set the values of the varibales
    def salary(self,value):
        if value>0:
            self._salary=value
        else:
            raise ValueError("Age cannot be deleted if 24 or above")
    @age.deleter
    def age(self):
        if self._age <24:
            del self._age
            print("Age is deleted")
        else:
            raise ValueError("Age cannot be deleted if 24 or above")
            
employee=Employee('Avinash',20,'Sr.QA',30000)
print(employee.name)
print(employee.salary)
employee.salary=40000
print(employee.salary)
del employee.age
```
