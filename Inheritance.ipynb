```python
'''
Inheritance -- This concept is similar to inheriting properties from parents if they have any in real world
in python from parent class we can inherit variables/instance varibles and methods 
This will help in reusing the code and reduces the ease of readabily and modification of the code
'''
class Animal: #This is a parent class
    def __init__(self,name,color): #constructer to define the instance varibles
        self.name=name
        self.color=color
    def eat(self): #method of the Animal class
        print(f"{self.name} is eating")
    def sleep(self): #method of the Animal class
        print(f"{self.name} is sleeping")
class Dog(Animal):
    def sound(self):#own methods of the child
        print(f"{self.name}: Barks!!!")
class Cat(Animal):
    def sound(self):#own methods of the child
        print(f"{self.name}: Meow!!!")
class Elephant(Animal):
    def sound(self):#own methods of the child
        print(f"{self.name}: taaaa!")
obj1=Dog('Rado','White')
obj2=Cat('Taco','Black')
obj3=Elephant('Bheema','Blue')
print(obj3.name)
print(obj3.color)
obj3.eat()
obj3.sleep()
obj3.sound()
```
