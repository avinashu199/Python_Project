Basic functions that are usefull in python are as mentioned below:
round -- using this we can round off the value to the whole number
abs-- absolute value is the distance between the actual value and zero
min-- smallest number in the given combination
max-- largest number in the given combination

```python
a=30.33
print(round(a))
```

```python
b=-3
print(abs(b))
```

```python
c=7
d=4
e=1
f=6
g=9
print(min(c,d,e,f,g))
print(max(c,d,e,f,g))
print(max(c,d))
```

Math functions -- usefull mathematical functions which we can use them by importing math module

```python
#power
import math
a=3
b=5
result=math.pow(b,a)
print(result)
```

```python
#pi
import math
print(math.pi)
```

```python
#ceiling and floor
import math
d=3.33
print(math.ceil(d))
print(math.floor(d))
```
