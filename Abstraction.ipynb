```python
'''
Abstraction - in python is like a blue print where we metion the methods which need to be followed or implemented in the child class.
Implementation details will not be there in the abstract class, only base strcuture will be there in the abstract class
'''
from abc import ABC,abstractmethod
class Shape(ABC): #this method now inherits the abstract class so we cannot create the object for this specific class
    @abstractmethod #this means that specific methods under this should be implemented and code also should be written in the child classes
    def area(self):
        pass #place holder
class Circle(Shape):
    def __init__(self,radius): #Defining constants using init constrctor
        self.radius=radius
    def area(self):#using the instance variables and defining the implementation of abstact method 
        return 3.14 *self.radius**2
class Square(Shape):
    def __init__(self,side):#Defining constants using init constrctor
        self.side=side
    def area(self):#using the instance variables and defining the implementation of abstact method
        return self.side*self.side
class Rectangle(Shape):
    def __init__(self,length,width):#Defining constants using init constrctor
        self.length=length
        self.width=width
    def area(self):#using the instance variables and defining the implementation of abstact method
        return self.length*self.width

circle=Circle(7)
square=Square(4)
rectangle=Rectangle(6,9)
print(circle.area())
print(square.area())
print(rectangle.area())
```
