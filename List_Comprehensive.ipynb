```python
'''
List comprehensive-- is like short hand if, but this is used to create the list in a better way with less number of code
'''
```

```python
'''
Short hand if 
syntax: code if condition else code
'''
number=30
"greater" if number>20 else "smaller"
```

```python
'''
List comprehensive:
syntax: [expression for iterator in iterable condition]
'''
#with list of numbers- without condition
numbers=[1,3,7,9]
new_numbers=[number*2 for number in numbers]
print(new_numbers)

#with list of words/strings- without condition
names=["Avinash","Naveen","Mahesh","Pavan"]
new_names=[name.upper() for name in names]
print(new_names)
```

```python
#more examples with list of numbers- without condition
numbers=[7,4,1,6,9]
new_numbers=[number*number for number in numbers]
print(new_numbers)
```

```python
#more examples with strings- without condition
bikes=["Hero","Honda","Yamoha"]
new_bikes=[bike.find('o')for bike in bikes]
print(new_bikes)
```

```python
#Complete example with expression ,for loop syntax and condition as well with numbers
numbers=[7,4,1,6,9]

#now create a list using comprehensive syntax where new list should only have even numbers from old list and those even numbers should be doubled

new_numbers=[number*2 for number in numbers if number%2==0]
print(new_numbers)
```

```python
#Complete example with expression ,for loop syntax and condition as well with strings
names=["Avinash","Ayansh","Kavita"]
new_names=[name.upper() for name in names if len(name)<=6]
print(new_names)
```
