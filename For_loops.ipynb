```python
#For loops are usefull- when we know how many times we would like to repated the loop. ex: If we want to repeat the loop for fixed amount of time then we can use for loop
#While loop is usefull when we want to loop for almost infinite amount of time
#Both in for loop and while loop- we can use break & continue statements. 
#Break- stop the execution
#Continue- will skip the execution
```

```python
#for loop syntax we can use string or range or list or tuple anything to loop that number of times
#write a program to print numbers between 1 to 10
for i in range(10): #we know the number of times to use the loop hence we are using for loop
    print(i)
```

```python
#same program using while loop
a=1
while a>0 and a<11:
    print(a)
    a=a+1
```

```python
#In for loop- instead of range we can use tuple list string anything that can be iterable
name='Ayaansh'
for i in name:
    print(i)
```

```python
#We can use both continue and break in for loop and while loop as well
name='Avinash Uppalapati'
for i in name:
    if i==' ':
        continue
    else:
        print(i)
```

```python
#break in for loop
name='Avinash'
for i in name:
    if i=='a':
        break
    else:
        print(i)
```

```python
'''
****
***
**
*
'''
lines=int(input('Enter the number of lines:'))
for i in range(lines):
    for j in range(lines):
        if j>= i:
            print('*',end=' ')
    print('')
```

```python
#combination of both programs
'''
*
**
***
****
***
**
*
'''
lines=int(input("Enter the number of lines you need"))
for i in range(lines):
    for j in range(lines):
        if j<=i:
            print('*',end=' ')
    print('')
for i in range(lines-1):
    for j in range(lines-1):
        if j>= i:
            print('*',end=' ')
    print('')
```
