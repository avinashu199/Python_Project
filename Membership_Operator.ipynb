```python
'''
Membership operator- In & not in. This operator can be used in all the iterables ex: String,List,Tuple,Set,Dictionary
'''
```

```python
#Membership operator in String
moviename="Salar"
guess=input("Guess the letter from prabhas latest movie:")
if guess in moviename:
    print(f"{guess} is correct")
else:
    print("You guess is wrong")
```

```python
#Membership operator in List
cars=["Ford","TATA","Mahendra"]
if "Ford" in cars:
    print("Car is found")
else:
    print("Car not found")
```

```python
#Membership operator in Set
bikes={"Hero","Honda","Yamaha"}
if "Honda" in bikes:
    print("Bike found in the set")
else:
    print("Bike not found in the set")
```

```python
#Membership operator in Tuple
movies=("Salar","Mirchi","Bahubali")
if "Mirchi" in movies:
    print("Movie found in the tuple")
else:
    print("Movie not found in the tuple")
```

```python
#Membership operator in Dictionary
bills={"Axis":"Neo","ICICI":"Amazon Pay","IDFC":"Digital"}
if "Neo" in bills.values():
    print("Card is there")
else:
    print("Card is not there")
```
