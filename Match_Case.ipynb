```python
'''
Match case is like switch statement in java.It is introduced from 3.10 version of python as we are currently using 3.12 we can use the statement.
switch is like alternative of using multiple if elif statements

'''
```

```python
#Traditional way of using 
day=int(input("Enter the number to select the corresponding day:"))
if day==1:
    print("Monday")
elif day==2:
    print("Tuesday")
elif day==3:
    print("Wednesday")
elif day==4:
    print("Thursday")
elif day==5:
    print("Friday")
elif day==6:
    print("Saturday")
elif day==7:
    print("Sunday")
else:
    print("Invalid- Enter between 1-7")
```

```python
#same we can use this in match case(same like switch case in java)
day=int(input("Enter the number to select the corresponding day:"))
match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
       print("Sunday")
    case _:
       print("Invalid- Enter between 1-7")
```

```python
#Another example for weekend
day=input("Enter the day:")
match day:
    case "Monday":
        print("Weekday")
    case "Tuesday":
        print("Weekday")
    case "Wednesday":
        print("Weekday")
    case "Thursday":
        print("Weekday")
    case "Friday":
        print("Weekday")
    case "Saturday":
        print("Weekend")
    case "Sunday":
        print("Weekend")
    case _:
        print("Re-enter correct day of the week:")
```

```python
#Reduce the lines of the code
day=input("Enter the day:")
match day:
    case "Monday" | "Tuesday" |"Wednesday"|"Thursday"| "Friday":
        print("Weekday")
    case "Saturday"|"Sunday":
        print("Weekend")
    case _:
        print("Re-enter correct day of the week:")
```
