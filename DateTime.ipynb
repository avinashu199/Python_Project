```python
'''
In python we have datetime module which has build in classes and methods using which we can play around with the dates and times
1.time method
2.date method
'''
```

```python
#working with time method
from datetime import time
new_time=time(8,30)
print(new_time)
```

```python
#working with date method
from datetime import date
new_date=date(2025,12,31)
print(new_date)

#we can also print the current date
current_date=date.today()
print(current_date)
```

```python
#to print current time and date we need to access datetime class inside datetime module
import datetime
current_time=datetime.datetime.now()
print(current_time)

#we can also print our own date and time using this datetimeclass
new_datetime=datetime.datetime(2025,12,31,12,0,0)
print(new_datetime)
```
