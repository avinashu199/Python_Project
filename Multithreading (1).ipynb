```python
'''
Multithreading-- is like a multitasking at the same time
ex: we can eat and work at the same time
Traditional way of programming what will happen is we will execute eating first and after that we will execute working

but using this concept of multithreading we can acheive multitasking at the same time
Inorder to do this we have to import Threading moudle first
then we have to load the task to different threads and start them-- to acheive this we have to use the Thread class where we need to
pass keyword arguments target=method name.
1.when methods have no arguments-- then no need only target keyword is enough
2.when methods are having arguments-- along with the target keyword we need to pass args another keyword but this accepts only tuple

****Note*****
1.We need to start the thread -- to run the code
2.we need to join the threads incase if we want to perform multi tasking first and other lines of code which is not part of multitasking
3.Threads that can be quickly completed will be executed first and others later on
'''
```

```python
import threading #this is the moudle of python using which we can perform multi tasking
import time #To make each methods time taking
#for multitasking create few methods- this time without arguments
def music():
    time.sleep(5)
    print("Avinash is listnening to music")
def eating():
    time.sleep(10)
    print("Avinash is eating food")
def working():
    time.sleep(2)
    print("Avinash is working")

#Traditional way to executing the methods
#music()
#eating()
#working()

#using multi threading we can execute the program simultaeniously
#for load the task into threads
task1=threading.Thread(target=music)
task1.start()
task2=threading.Thread(target=eating)
task2.start()
task3=threading.Thread(target=working)
task3.start()
#Join should be used to see the program executed correctly, if we dont use this we cannot see the results as before printing results pyton will move to the other method
#so we cannot see the results , to see the results we should use join , so python will tell that wait till we finish each task and move to other task after finishing one after other
task1.join()
task2.join()
task3.join()
```

```python
#incase if the method is having arguments
#use the same program
import threading #this is the moudle of python using which we can perform multi tasking
import time #To make each methods time taking
#for multitasking create few methods- this time with arguments
def music(song,artist):
    time.sleep(5)
    print(f"Avinash is listnening to {song} sung by {artist}")
def eating(food):
    time.sleep(10)
    print(f"Avinash is eating {food}")
def working():
    time.sleep(2)
    print("Avinash is working")


#using multi threading we can execute the program simultaeniously
#for load the task into threads
task1=threading.Thread(target=music,args=("Kingdom","Anirudh"))
task1.start()
task2=threading.Thread(target=eating("Biryani",))
task2.start()
task3=threading.Thread(target=working)
task3.start()
#Join should be used to see the program executed correctly, if we dont use this we cannot see the results as before printing results pyton will move to the other method
#so we cannot see the results , to see the results we should use join , so python will tell that wait till we finish each task and move to other task after finishing one after other
task1.join()
task2.join()
task3.join()
```
