```python
'''
We have magic methods, these magic methods are mainly useful for us to play around with the objects of the class
1) __init__ we can initialize the instance variables in the class
2)__str__ we can modify the result when we call the object which means we can print what we want to print instead of memory location
#We can also perform a bit of arithmetic operations with objects of the classes using the special methods as below
3)__eq__ we can compare the objects using this magic method,
but we should pass 2 parameters which are self and keyword. using these 2 pointers we can write a condition to compare
4)__lt__ we can compare the objects using this magic method,
but we should pass 2 parameters which are self and keyword. using these 2 pointers we can write a condition to compare
5)__gt__we can compare the objects using this magic method,
but we should pass 2 parameters which are self and keyword. using these 2 pointers we can write a condition to compareo[[,k
6)__add__we can add the values directly using the objects using 2 pointers using this magic method
7)__contains__ we can check a particular work is there in the instance varibales or not using the object of the class directly.
8)__getitem__
'''
```

```python
#1) __init__ method
class Book:
    def __init__(self,name,author,pages): #initialise these instance varibales
        self.name=name
        self.author=author
        self.pages=pages
#create an object for the class and try to call the object
book1=Book("2 States","Chetan Bhagat",120)
book2=Book("The God of Small Things","Arundhati Roy",150)
book3=Book("Ramayana Vishavruksham","Ranganayakamma",300)

#If we call the object book then it will display the memory location of the book
print(book3)
```

```python
'''
As mentioned obove, instead of memory location we can print/customise our own stuff using __str__ method in python
'''
#2) __str__ method
class Book:
    def __init__(self,name,author,pages): #initialise these instance varibales
        self.name=name
        self.author=author
        self.pages=pages
    def __str__(self):
        return f"Great book '{self.name}' by {self.author}"
#create an object for the class and try to call the object
book1=Book("2 States","Chetan Bhagat",120)
book2=Book("The God of Small Things","Arundhati Roy",150)
book3=Book("Ramayana Vishavruksham","Ranganayakamma",300)

print(book1)
print(book2)
print(book3)
```

```python
'''
We can perform arithmetic opertaions on the objects as well if we use these magic methods
'''
print(book1>book2)
```

```python
#3) __eq__ method
class Book:
    def __init__(self,name,author,pages): #initialise these instance varibales
        self.name=name
        self.author=author
        self.pages=pages
    def __eq__(self,keyword):
        return True if self.name==keyword.name and self.author==keyword.author else False
#create an object for the class and try to call the object
book1=Book("2 States","Chetan Bhagat",120)
book2=Book("2 States","Chetan Bhagat",220)
#book2=Book("The God of Small Things","Arundhati Roy",150)
book3=Book("Ramayana Vishavruksham","Ranganayakamma",300)

print(book1==book2)
```

```python
#4) __lt__ method
class Book:
    def __init__(self,name,author,pages): #initialise these instance varibales
        self.name=name
        self.author=author
        self.pages=pages
    def __lt__(self,keyword):
        return True if self.pages<keyword.pages else False
#create an object for the class and try to call the object
book1=Book("2 States","Chetan Bhagat",120)
book2=Book("The God of Small Things","Arundhati Roy",150)
book3=Book("Ramayana Vishavruksham","Ranganayakamma",300)

print(book1<book2)
```

```python
#5) __gt__ method
class Book:
    def __init__(self,name,author,pages): #initialise these instance varibales
        self.name=name
        self.author=author
        self.pages=pages
    def __gt__(self,keyword):
        return True if self.pages>keyword.pages else False
#create an object for the class and try to call the object
book1=Book("2 States","Chetan Bhagat",120)
book2=Book("The God of Small Things","Arundhati Roy",150)
book3=Book("Ramayana Vishavruksham","Ranganayakamma",300)

print(book3>book2)
```

```python
#6) __add__ method
class Book:
    def __init__(self,name,author,pages): #initialise these instance varibales
        self.name=name
        self.author=author
        self.pages=pages
    def __add__(self,keyword):
        return self.pages+keyword.pages
#create an object for the class and try to call the object
book1=Book("2 States","Chetan Bhagat",120)
book2=Book("The God of Small Things","Arundhati Roy",150)
book3=Book("Ramayana Vishavruksham","Ranganayakamma",300)

print(book1+book2)
```

```python
#7)__contains__method
class Book:
    def __init__(self,name,author,pages): #initialise these instance varibales
        self.name=name
        self.author=author
        self.pages=pages
    def __contains__(self,keyword):
        return True if keyword in self.name or keyword in self.author else False
#create an object for the class and try to call the object
book1=Book("2 States","Chetan Bhagat",120)
book2=Book("The God of Small Things","Arundhati Roy",150)
book3=Book("Ramayana Vishavruksham","Ranganayakamma",300)

print("God" in book1)
```

```python
#8)__getitem__ it is more likely to convert the object in list,tuple,set etc
class Book:
    def __init__(self,name,author,pages): #initialise these instance varibales
        self.name=name
        self.author=author
        self.pages=pages
    def __getitem__(self,index):
        return self.name[index]
#create an object for the class and try to call the object
book1=Book("2 States","Chetan Bhagat",120)
book2=Book("The God of Small Things","Arundhati Roy",150)
book3=Book("Ramayana Vishavruksham","Ranganayakamma",300)

print(book1[0])
```
