```python
#slicing and negative slicing 
#syntax-- start:stop:step
```

```python
card='4315-5168-0516-2877'
print(card[::]) #if start,stop,step are not mentioned then all will be printed end to end
#Focusing on start
#Print only items using start
print(card[5::])
#we can use stop also but it is not inclusive
print(card[10:14:])
#we can also use step to print alternative wise aswell
print(card[::2])
#if we use -1 then reverse will be printed
print(card[::-1])
```

```python
#negative slicing we can use negative values for start and stop as well
name='avinash uppalapati'
#if we want to print last 4 letters then we can use negative slicing
print(name[-4::])
#we can use same things using negative slicing
print(name[-10:-1:])
#we can use negative values in the step value as well
print(name[::-2])
#if we want to print first 7 letters then we can use negative slicing
print(name[6::-1])
```
