```python
'''
Exception handling-- Like try catch block using which we can handle any exceptions caused while executing the program
we have few things in python
try-- it will execute this code first
except-- this code will be executed incase of any exceptions
finally -- no matter what this block of code will be executed
'''
```

```python
try:
    num=int(input("Enter the number:"))
    1/num
except ZeroDivisionError:
    print("Please don't enter 0 which is causing this issue:")
except ValueError:
    print("Please enter a valid whole number except 0")
finally:
    print("There are exception in this program")
```

```python
#Another away of writing simple try except finally block is
try:
    num=int(input("Enter the number:"))
    1/num
except Exception:
    print("Bulk way of handling all different types of exception")
finally:
    print("By default execution of code block")
```
