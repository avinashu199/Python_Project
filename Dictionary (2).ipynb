```python
'''
Dictionary is also one of the data type like list,tuple, set etc etc

syntax: {"Keys":"Values"}

We have some methods that can be used for playing around with the dictionaries like we have the methods for list and string
1.get method- [Only keys are accepeted in this method]This method will take one attribute which is 'key' and will return the corresponding value for that specific key.
2.values method- will display only the values of the dictionary
3.keys method- will display only keys of the dictionary 
4.update method- this will help us in updating the values in the list, we can add new values as well using this same method
5.pop method-This will remove the specific values of the dictionoary, we only need to pass the specific key which we want to remove
6.clear method-This will remove the whole dictionary
'''
```

```python
college={"CSE":"Python","ECE":"Diodes & Circuits","MEC":"MAT Lab"}
print(college)
```

```python
print(help(college))
```

```python
#get method
result=college.get("CSE")
print(result)
```

```python
#values method
result=college.values()
print(result)
```

```python
#keys method
result=college.keys()
print(result)
```

```python
#update method
college={"CSE":"Python","ECE":"Diodes & Circuits","MEC":"MAT Lab"}
print(college)

#college.update({"MEC":"Survey"})
#print(college)

#we can add all new values together in the same dictionary using the same update method
college.update({"Civil":"Survey"})
print(college)
```

```python
#pop method
college.pop("ECE")
print(college)
```

```python
#copy method becuase we cannot store the updated dictionary into new once we can copy and after we can update the dictionary
college={"CSE":"Python","ECE":"Diodes & Circuits","MEC":"MAT Lab"}
print(college)
#we can add all new values together in the same dictionary using the same update method
college.update({"Civil":"Survey"})
pvpsit=college.copy()
print(pvpsit)
```

```python
#clear method
college.clear()
print(college)
```
