```python
'''
Revisiting 3 main topics of the class
1.Instance variables & Instance Methods --- anything that is related to object creation and calling methods based on the objects
2.Static Methods[@staticmethod]-- Independent methods and have no parameters as self and no need to any object creation, we can directly call that method using class name
3.Class Methods[@classmethod]--if we need to work with the class varibales then we can use this concept, this classmemthds will use cls as parameter
'''
```

```python
#class methods[@classmethod] & cls as parameter
class Pvpsit:
    count=0 #create a class variable
    total=0 #create a class variable
    #create an instance varibale and instance method
    def __init__(self,name,gpa):
        self.name=name
        self.gpa=gpa
        Pvpsit.count=Pvpsit.count+1
        Pvpsit.total=Pvpsit.total+gpa
    def student(self):
        print(f"{self.student}={self.gpa}")
    #now use those class variables and work on them to use them we have to create a class method
    @classmethod
    def average(cls):
        if cls.count==0:
            return 0
        else:
            return f"{cls.total/cls.count}"
stud1=Pvpsit("Avinash",3.3)
stud2=Pvpsit("Kavita",4.0)
stud3=Pvpsit("Ayaansh",1.7)
print(Pvpsit.average())
```
