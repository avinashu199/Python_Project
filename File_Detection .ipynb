```python
#we can also use the OS module to understand if the location is having a file or location is just a directory

#Using OS to undertsnad if it is a file or not
import os 
file_path="C:\\Users\\desai\\OneDrive\\Desktop\\Resume-Avinash" #this is not a file but it is just a directory location
print(os.path.isfile(file_path))

#hence it is not file ccheck if it is a directory or not
print(os.path.isdir(file_path))
```
