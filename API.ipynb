```python
'''
We can handle API using requests module in python
'''
import requests #using this requests module we can work with the API's in python

base_url="https://pokeapi.co/api/v2/pokemon/"# this will be the url which base, but based on our pokemon url will keep changing at the end
pokemon_name="palkia"
updated_url=f"{base_url}{pokemon_name}"
def pokemon_details(pokemon_name):
    response=requests.get(updated_url) #will check the API available or not
    status=response.status_code #This will store the status code from the response of the api get method
    if status==200:
        result=response.json()
        return result
    else:
        print(f"{pokemon_name} is not available in the API")
        
output=pokemon_details(pokemon_name)
print(f"Name:{output['name']}")
print(f"Height:{output['height']}")
print(f"Weight:{output['weight']}")
```
