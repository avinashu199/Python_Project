```python
'''
in python we have CRUD operatorions which we can perform on the files
C-Creation
R-Reading
U-Updating
D-Deleting
'''
```

```python
'''
for TXT and Json files we have similar approach which means we have file modes which need to be specified
if we want to just read the existing file then we have to use the mode as R while opening the file
'''
```

```python
#TXT files
'''
1.Read the existing file- for that we need to use open method which will ask for 2 parameteres 1= path/file name , 2= mode of viewing
This will return the file object as output which we will be using to print to see the items of the file
'''
file_name="note.txt"
output=open(file_name,"r")
print(output.read())
output.close()
```

```python
'''
write mode- this will create the file incase if the file name is not there in the location, or if it is there then it will replace the content
'''
file_name="C:\\Users\\desai\\OneDrive\\Desktop\\Ayaansh.txt"
obj=open(file_name,'w')
obj.write("Kavita is a good girl")
obj.close()
```
