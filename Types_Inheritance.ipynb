```python
'''
There are 2 types of inheritance 
1.Multiple Inheritance-- single child class can inherit the instance variables and methods from multiple parent class
2.Multilevel Inheritance-- single child class can inheirt from another child class as well this is like father inherits from grandfather and son now inherts from the father 
'''
```

```python
#Multiple Inheritance
class Animal:#parent class
    def desca(self):
        print("I'm an animal with 2 legs and I can bite")
class Bird:#parent class
    def descb(self):
        print("I'm a bird with wings and I can fly")
class Tiger(Animal):#child class
    pass
class Hen(Animal,Bird):#Multiple inheritance
    pass
obj1=Tiger()
obj2=Hen()
obj1.desca()
obj2.descb()
obj2.desca()
```

```python
#Multilevel Inheritance
class Grandfather: #Parent class
    def agri(self):
        print("10 acers of land")
class Father(Grandfather): #child class
    def living(self):
        print("1 aprtments in hyd")
class Son(Father): #Multi level inheritance
    def stocks(self):
        print("Invested in Multual funds")
obj1=Son()
obj1.agri()
obj1.living()
obj1.stocks()
```
