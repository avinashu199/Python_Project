```python
#importing modules
#Inbuilt modules
print(help("modules"))
```

```python
#how to import the modules
'''
There are 3 ways to import modules
1.import modulename
2.import modulename as aliasname
3.from modulename import function/variable name
'''
#importing using the module name
import math
print(math.pi) #we are accessing the constants or variables that are available in the math module, so we can print them directly
print(math.e) #we are accessing the constants or variables that are available in the math module, so we can print them directly

'''
To access the functions of the module, we need to first understand the funchions 
which means we need to understand whether function accepts any arguments/parameters or how many arguments/parameters does that function
accepts. we need print that because function returns certain values, whatever it returns will be printed in the console based on the
arguments/parameters which we are passing
'''

print(math.sqrt(9))
print(math.pow(2,3))
```

```python
#importing using alias name
import math as d
print(d.pi)
print(d.pow(3,3))
```

```python
#importing only function/constant directly
from math import e
print(e)

from math import pow
print(pow(2,5))
```
