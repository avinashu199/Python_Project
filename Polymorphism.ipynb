```python
#Inheritance
'''
Polymorphism- many forms 
Using inheritance if we create an object for child class then that object has 2 forms--- 1 form is child class(which means the details of the child class)
2 form is parent class(same oject now have a different face using which we can access the elements of the parent class)
'''
class Shape:#parent class
    def __init__(self,radius):
        self.radius=radius
class Circle(Shape):#child class
    def area(self):
        return 3.14*self.radius*self.radius
class Square(Shape):#child class
    def area(self):
        return self.radius*self.radius
class Rectangle(Shape):#child class
    def __init__(self,length,width):
        self.length=length
        self.width=width
    def area(self):
        return self.length*self.width
#now create an object for square
square=Square(5)
#this object has 2 faces --- 
#same object is used to access the square class
#Same object has 2nd face using which we can access the parent class i.e Shape
```
