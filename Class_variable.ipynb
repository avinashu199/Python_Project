```python
'''
Python when we are working on class, we need to understad the concept of instance variable and class variable
'''
```

```python
#instance variable
class Hero:
    def __init__(self,name,industry,movie):
        #below are the instance variables of this class and they are specific to the object and cannot be generalised
        self.name=name
        self.industry=industry
        self.movie=movie
#So these instance variables are specific to this particular object
obj1=Hero('Prabhas','Tollywood','SPIRIT')
print(obj1.name)
print(obj1.industry)
print(obj1.movie)
#Now new object hence those old valriables of obj1 is not considered
obj2=Hero('Karthi','Kollywoood','Satyam Sundaram')
print(obj2.name)
print(obj2.industry)
print(obj2.movie)
```

```python
'''
class variables are created outside the constructor and methods, hence we can access the class varibales irrespective of the object created or not.
class varibales can be remainted common across the whole class and no need to create object to access the class variable
'''
class Bike:
    best_bike='Karizma'
    def __init__(self,name,cc,color):
        self.name=name
        self.cc=cc
        self.color=color
print(Bike.best_bike) #we can call the class variable directly without using any of the object. DIrectly we can use the class name and call the object
obj1=Bike('Pulsar',150,'Black')
print(obj1.name)
print(obj1.cc)
print(obj1.color)
print(obj1.best_bike) #we can call class variable using object as well but it is recommended to call the class variables using the class name directly
```
