```python
'''
Static method- is used when we dont depend anything on the instance variables and instance methods, 
hence we dont need object creation and object refereces and this static method is independent and doesnot require any object creation
This much like class variables we can access them using class name
'''
```

```python
class Avinash:
    boy='Ayaansh'
    def __init__(self,job,salary): #Constructor
        self.job=job #Instance variable
        self.salary=salary #Instance variable
    def professional_life(self): #Instance method
        print(f"I'm working as {self.job} with ${self.salary}")
        
    @staticmethod
    def personal_life():
        print(f"Im father of a {Avinash.boy} and happiely married for 2 years")

#to access static method we dont need to creat an object , we can access directly using the class name
Avinash.personal_life()

#To access instace method and instance variables we need to create an object for the class
obj=Avinash("Senior Quality Analyst",85000)
obj.professional_life()
```
