String methods are very usefull to workaround with the strings. This concept is very much usefull when we attend interviews
1.legth method -- this is generic method to be used to identify the length of the string
2.find method-- this is used to identify the index position of the specific element in the given string.
3.upper method- make the whole string in capital letters
4.lower method- make the whole string in lower letters
5.isapla- returns true only if alphabets are there else it will return false
6.isdigit- returns true only if there are digits else it will return false
7.count-- method return the integer value of the total number of time a specific word is repeated, if none is repeated then it will display 0
8.capitalize- This will only make the first alphabet a capital letter
9.replace- will help in replacing the specified letter

```python
name='Ayaansh'
length=len(name)
print(length)
```

```python
name='Avinash Uppalapati'
result=name.find('_')
print(result)
```

```python
name='Avinash Uppalapati'
result=name.upper()
print(result)
```

```python
name='Avinash Uppalapati'
result=name.lower()
print(result)
```

```python
name='Rangaraju123'
result=name.isalpha()
print(result)
```

```python
pincode='520011'
result=pincode.isdigit()
print(result)
```

```python
name='Rangaraju11'
result=name.count('1')
print(result)
```

```python
name='rangaraju'
result=name.capitalize()
print(result)
```

```python
name='rangaraju'
result=name.replace('a','A')
print(result)
```

```python
name='rangaraju'
result=name.replace(name[0],'R')
print(result)
```
