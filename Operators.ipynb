```python
#Arithmatic operators i.e + - * / % // **

a=float(input("Enter the first number"))
b=float(input("Enter the second number"))
sum= a+b
sub=a-b
mul=a*b
div=a/b
mod=a%b
dmul= a**b
ddiv=a//b
print(f"The sum of two numbers is:{sum}")
print(f"The sub of two numbers is:{sub}")
print(f"The mul of two numbers is:{mul}")
print(f"The div of two numbers is:{div}")
print(f"The mod of two numbers is:{mod}")
print(f"The dmul of two numbers is:{dmul}")
print(f"The ddiv of two numbers is:{ddiv}")
```

```python
#Assignment operator
num=int(input("Enter a number:"))
#we can use the same arithmetic in a different format
num+=1 #num=num+1
print(f"Add value is :{num}")
num-=1#num=num-1
print(f"Sub value is :{num}")
num*=2#num=num*2
print(f"Mul value is :{num}")
num/=2#num=num/2
print(f"Div value is :{num}")
```

```python
#comparision operators < > <= >= != ==
a=int(input("Enter the number1:"))
b=int(input("Enter the number2:"))
print(a<b)
print(a>b)
print(a<=b)
print(a>=b)
print(a==b)
print(a!=b)
```

```python
#logical operator -- these are mainly used between two conditions which are OR ,AND, NOT
a=3033
b=3113
#OR operator
if a<b or a==b:
    print("A is second rank holder")
else:
    print("A is the topper")
#AND operator
if a<b and a%2==1:
    print("This is my favourate number")
else:
    print("This is not my favourate number")
#NOT operator
if not(a<b and a%2==1):
    print("This is my favourate number")
else:
    print("This is not my favourate number")
```

```python
#Membership operator- In & not in- this is mostly applicable for list 
a=[7,4,1,6,9,3]
print(8 not in a)
```

```python
#Identical Operators - IS,ISNOT both memory location should be same then only output will be true if not output will be false
#Memory locations are different
a=3033
b=3033
print(a is not b)

c=3113
d=c
print(c is d)
print(c is not d)
```
